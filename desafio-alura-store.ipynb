### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

#0. Valores nulos

```python
lojas = [loja, loja2, loja3, loja4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
dfs_dados = {}

for i in range(len(lojas)):
    dados_nulos = lojas[i].isnull().sum()

    df_dados_nulos = pd.DataFrame({'Dados': dados_nulos.index, 'Quantidades': dados_nulos.values})
    dfs_dados[nomes[i]] = df_dados_nulos

    print(f'{nomes[i]}\n')
    print(f'{df_dados_nulos}\n\n')
```

#1. Análise do faturamento


```python
lojas = [loja, loja2, loja3, loja4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = []
dfs_faturamento = {}


for i in range(len(lojas)):
    faturamento_total = lojas[i]['Preço'].sum()
    faturamentos.append(faturamento_total)

    df_faturamento = pd.DataFrame({'Loja': [nomes[i]], 'Faturamento Total (R$)': [faturamento_total]})
    dfs_faturamento[nomes[i]] = df_faturamento

    print(f'\n{df_faturamento}')
```

#1.1. Gráfico do Faturamento

```python
import matplotlib.pyplot as plt

print(f'\n \n')
plt.figure(figsize=(14,7))
plt.bar(nomes, faturamentos, color='red')
plt.xlabel('Loja', fontsize=15)
plt.ylabel('Faturamento Total (R$)', fontsize=15)
plt.title('Faturamento por Loja', fontsize=25)
plt.show()
```

# 2. Vendas por Categoria


```python
import matplotlib.pyplot as plt

lojas = [loja, loja2, loja3, loja4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
dfs_categorias = {}

for i in range(len(lojas)):
    produtos_vendidos = lojas[i]['Categoria do Produto'].value_counts()

    df_categorias = pd.DataFrame({'Categoria': produtos_vendidos.index, 'Quantidade': produtos_vendidos.values})

    dfs_categorias[nomes[i]] = df_categorias

    print(f'{nomes[i]}\n')
    print(f'{df_categorias}\n')
    print(f'Vendas totais: {produtos_vendidos.sum()}\n\n')
```

#2.1 Gráficos de Vendas por Categoria


```python
for i in range(len(lojas)):
    produtos_vendidos = lojas[i]['Categoria do Produto'].value_counts()

    plt.figure(figsize=(6,6))
    plt.pie(produtos_vendidos, labels=produtos_vendidos.index.str.capitalize(), autopct='%1.2f%%', startangle=90)
    plt.title(f'{nomes[i]} - Vendas Por Categoria')
    plt.show()
    print('\n\n')
```

# 3. Média de Avaliação das Lojas

```python
lojas = [loja, loja2, loja3, loja4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = []
dfs_medias = {}

for i in range(len(lojas)):
    m_avaliacao = round((lojas[i]['Avaliação da compra'].sum()) / (len(lojas[i]['Avaliação da compra'])), 2)
    medias.append(m_avaliacao)

    df_medias = pd.DataFrame({'Loja': [nomes[i]], 'Avaliação Média': [m_avaliacao]})
    dfs_medias[nomes[i]] = df_medias

    print(f'\n{df_medias}')
```

#3.1. Gráfico das Médias de Avaliações

```python
print(f'\n \n')
plt.figure(figsize=(14,7))
plt.plot(nomes, medias, marker='o', linestyle='-', color='red')
plt.xlabel('Loja', fontsize=15)
plt.ylabel('Avaliação Média', fontsize=15)
plt.title('Avaliação Média por Loja', fontsize=25)
plt.grid(True)
plt.show()
```

# 4. Produtos Mais Vendidos

```python
import matplotlib.pyplot as plt

lojas = [loja, loja2, loja3, loja4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
dfs_mais_vendidos = {}

for i in range(len(lojas)):
  mais_vendidos = lojas[i]['Produto'].value_counts().head(10)

  df_mais_vendidos = pd.DataFrame({'Produto': mais_vendidos.index, 'Quantidade': mais_vendidos.values})
  dfs_mais_vendidos[nomes[i]] = df_mais_vendidos

  print(f'Os 10 Produtos MAIS Vendidos - {nomes[i]}\n')
  print(f'{df_mais_vendidos}\n\n\n')

```

#5. Produtos Menos Vendidos

```python
import matplotlib.pyplot as plt

lojas = [loja, loja2, loja3, loja4]
dfs_menos_vendidos = {}

for i in range(len(lojas)):
  menos_vendidos = lojas[i]["Produto"].value_counts().nsmallest(10)

  df_menos_vendidos = pd.DataFrame({'Produto': menos_vendidos.index, 'Quantidade': menos_vendidos.values})
  dfs_menos_vendidos[nomes[i]] = df_menos_vendidos

  print(f'Os 10 Produtos MENOS Vendidos - {nomes[i]}\n')
  print(f'{df_menos_vendidos}\n\n\n')
```

# 6. Custo Médio do Frete por Loja


```python
import matplotlib.pyplot as plt

lojas = [loja, loja2, loja3, loja4]
fretes = []
dfs_fretes = {}

for i in range(len(lojas)):
  valor_frete = round(lojas[i]['Frete'].sum() / len(lojas[i]), 2)
  fretes.append(valor_frete)

  df_fretes = pd.DataFrame({'Loja': [nomes[i]], 'Frete Médio (R$)': [valor_frete]})
  dfs_fretes[nomes[i]] = df_fretes

  print(f'{df_fretes}\n')
```

#6.1. Gráfico do Custo do Frete

```python
plt.figure(figsize=(14,7))
plt.barh(nomes, fretes, color='blue')
plt.xlabel('Valor (R$)', fontsize=15)
plt.ylabel('Loja', fontsize=15)
plt.title('Valor do Médio Frete', fontsize=25)
plt.show()
```
